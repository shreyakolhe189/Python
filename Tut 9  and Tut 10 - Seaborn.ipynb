### Correlation with heatmap
- uses colored cells in monochromatic scale to show 2d correlation matrix between 2 discrete dimensions, very imp in feature selection

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

### Joinplot
- allows to study relationship between 2 numeric variables 
- central chart display their correlation
- usually a scatterplot, hexbin plot, 2D histogram, 2D density plot

#### Univariate Analysis - 2 features


```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex') #hexbin
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='scatter')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg') #regression
```

### Pairplot  
- also known as scatterplot
- one variable in same data row is matched with another variable's value
- shows all variables paired with all other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='species')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

### Distplot
- check distribution of columns features

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Categorical Plots
- boxplot
- violinplot
- countplot
- barplot

### countplot

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

### barplot

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

### boxplot

```python
sns.boxplot('smoker','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

### violinplot
- helps to see both distribution of data in terms of Kernel density estimation kde and boxplot

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
