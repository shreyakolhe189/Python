### Matplotlib
- plotting library for python & Numpy
- provides object oriented API for embedding plots into applications using general purpose GUI toolkits like Tkinter wxPython Qt GTK+

#### Pros :
- easy to get started for simple plots
- support custom labels and texts
- great control of every element in figure
- high quality output
- customizable

```python
import matplotlib.pyplot as plt

%matplotlib inline #if this then no need to write plt.show()
```

```python
x = np.arange(1,11)
y = 3 * x + 5

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Matplotlib Demo")
plt.plot(x,y)
```

```python
x = np.arange(0,4*np.pi,0.1) # np.pi = 3.14...
y = np.sin(x)

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("sin wave form")
plt.plot(x,y,c="r")
```

```python
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,2,1)
plt.plot(x,y_sin)
plt.title("Sine")

plt.subplot(2,1,2)
plt.plot(x,y_cos,"g--")
plt.title("Cosine")
```

## Bar Plot

```python
x = [2,8,10]
y = [11,16,9]

x2 =[3,9,11]
y2 = [6,15,7]

plt.bar(x,y,color="y")
plt.bar(x2,y2,color="g")

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Bar Graph")

plt.show()
```

## Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a) #(bins=20) #bins - density of bin size
plt.title("Histogram")
plt.show()
```

## Box Plot


```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=False,patch_artist=False);
```

## Pie Charts

```python
labels = "python","c++","ruby","java"
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','pink']
explode = (0.1,0,0,0)
```

```python
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```
