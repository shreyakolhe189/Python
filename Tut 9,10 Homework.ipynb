### Iris

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
sns.heatmap(df.corr())
```

## Joinplot

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='hex')
```

```python
sns.jointplot(x='petal_length',y='petal_width',data=df,kind='reg')
```

## Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sepal_length')
```

```python
df['sepal_length'].value_counts
```

## Distplot

```python
sns.distplot(df['sepal_length'])
```

```python
sns.distplot(df['sepal_length'],kde=False,bins=10)
```

## Countplot

```python
sns.countplot('sepal_length',data=df)
```

## Barplot

```python
sns.barplot(x='sepal_length',y='petal_length',data=df)
```

## Boxplot

```python
sns.boxplot('sepal_length','petal_length',data=df,palette='rainbow')
```

```python
sns.boxplot(x='sepal_length',y='petal_length',hue='sepal_width',data=df)
```

## Violinplot

```python
sns.violinplot('sepal_length','petal_length',data=df,palette='rainbow')
```
